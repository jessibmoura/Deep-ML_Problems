```python
vectors = [[1, 2, 3], [4, 5, 6]]
```

```python
import numpy as np

def calculate_covariance_matrix(vectors: list[list[float]]) -> list[list[float]]:
    feats = len(vectors)
    obs = len(vectors[0])
    covar_matrix = covariance_matrix = np.zeros((feats,feats))

    means = [np.mean(feature) for feature in vectors]

    for i in range(feats):
        for j in range(i, feats):
            covariance = sum((vectors[i][k] - means[i]) * (vectors[j][k] - means[j]) for k in range(obs)) / (obs - 1)
            covar_matrix[i][j] = covar_matrix[j][i] = covariance

    return covariance_matrix.tolist()
```

```python
calculate_covariance_matrix(vectors)
```
