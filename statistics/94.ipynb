```python
x = [1, 1, 0, 0]
y = [0, 0, 1, 1]
```

```python
def phi_corr(x: list[int], y: list[int]) -> float:
    x_00 = x_01 = x_10 = x_11 = 0
    for i in range(len(x)):
        if x[i] == 0 and y[i] == 0:
            x_00 += 1
        elif x[i] == 0 and y[i] == 1:
            x_01 += 1
        elif x[i] == 1 and  y[i] == 0:
            x_10 += 1
        elif x[i] == 1 and  y[i] == 1:
            x_11 += 1

    numerator = (x_00 * x_11) - (x_01 * x_10)
    denominator = ((x_11 + x_01) * (x_11 + x_10) * (x_00 + x_01) * (x_00 + x_10)) ** 0.5

    if denominator == 0:
        return 0.0
    
    val = numerator / denominator
    return round(val,4)
```

```python
print(phi_corr([1, 1, 0, 0], [0, 0, 1, 1]))
```

```python
print(phi_corr([1, 1, 0, 0], [1, 0, 1, 1]))
```
