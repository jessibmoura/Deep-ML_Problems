```python
a = [[1,2],[2,4]]
b = [1,2]
```

```python
def matrix_dot_vector(a:list[list[int|float]],b:list[int|float])-> list[int|float]:
    c = []
    if len(a[0]) == len(b):
        for i in a:
            t = 0
            for j in range(0,len(i)):
                t += i[j] * b[j]
            c.append(t)
        return c
    return -1
```

```python
matrix_dot_vector(a,b)
```
