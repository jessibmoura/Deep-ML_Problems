```python
a = [[1,2,3],[4,5,6]]
```

```python
def transpose_matrix(a: list[list[int|float]]) -> list[list[int|float]]:
    b = []
    new_row = []
    for col in range(len(a[0])):
        new_row = []
        for row in range(len(a)):
            new_row.append(a[row][col])
        b.append(new_row)
    return b
```

```python
transpose_matrix(a)
```
