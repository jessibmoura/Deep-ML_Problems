```python
B = [[1, 0, 0],[0, 1, 0],[0, 0, 1]]
C = [[1, 2.3, 3],[4.4, 25, 6],[7.4, 8, 9]]
```

```python
import numpy as np

def transform_basis(B: list[list[int]], C: list[list[int]]) -> list[list[float]]:
	P = np.linalg.inv(C).dot(B)
	P = np.round(P,4)
	return P
```

```python
transform_basis(B,C)
```
