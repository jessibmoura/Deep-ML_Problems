```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
mode = 'column'
```

```python
def calculate_matrix_mean(matrix: list[list[float]], mode: str) -> list[float]:
    means = []
    match mode:
        case 'column':
            for j in range(len(matrix[0])):
                total = 0
                for i in range(len(matrix)):
                    total += matrix[i][j]
                total = total/len(matrix[0])
                means.append(total)
        case 'row':
            for i in range(len(matrix)):
                total = 0
                for j in range(len(matrix[0])):
                    total += matrix[i][j]
                total = total/len(matrix[0])
                means.append(total)
    return means
```

```python
calculate_matrix_mean(matrix,mode)
```
